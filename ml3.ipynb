```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('D:\ml\exl\ml3onehotencoding.csv')
df
```

```python
df=df.dropna()
df
```

```python
dum=pd.get_dummies(df)
dum
```

```python
dumm=dum.drop(["Car Model_Mercedez Benz C class"],axis='columns')
```

```python
dumm
```

```python
X=dumm.drop('Sell',axis='columns')
X
```

```python
y=dum.Sell
y
```

```python
reg=LinearRegression()
reg.fit(X,y)
```

```python
reg.predict(X)
```

```python
reg.predict([[45000,4.0,0,0]])
```

```python
reg.predict([[8600,7,0,1]])
```

```python
reg.score(X,y)
```
